```python
import pandas as pd
import os
directory = r'C:\Users\mmohdat\Documents\ARTS\July-23\Vulcan ISM - Outages - [EU4] 2043\Export file'
def get_latest_file_df(mp):
    if os.path.exists(directory):
        files = os.listdir(directory)
        if len(files) > 0:
            # Sort the files by modification time (most recent first)
            sorted_files = sorted(files, key=lambda x: os.path.getmtime(os.path.join(directory, x)), reverse=True)

            latest_file = sorted_files[0]
            print("Latest file:", latest_file)
            df = pd.read_csv(directory + '\\' + latest_file)

            l = []

            for i in range(len(df)):
                l.append(mp)

            df['MP'] = l

            df.to_csv(os.getcwd() + "\\Export file\\" + latest_file, index=False)
            print(df)
        else:
            print("No files found in the directory.")
    else:
        print("Directory does not exist.")
```
